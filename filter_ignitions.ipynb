```python
f = open("data/ids.txt", "r")
names = []
for id in f:
    id = id.strip()
    names.append(id)
```

```python
import pandas as pd
full_ignitions_file_path = "data/yosemite_ignitions_firesize_100ac.csv"
filtered_ignitions_file_path = "data/yosemite_sample_ignitions.csv"

full_ignitions_df = pd.read_csv(full_ignitions_file_path)
filtered_ignitions_df = full_ignitions_df[(full_ignitions_df.filename.isin(names))]
filtered_ignitions_df.to_csv(filtered_ignitions_file_path, index=False)
filtered_ignitions_df
```
